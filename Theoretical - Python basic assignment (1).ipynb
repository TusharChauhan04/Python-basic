#Python basics Questions

1. What is Python, and why is it popular?

>> Python is a high-level, interpreted and dynamicall typed general-purpose programming language known for its simple, readable syntax, and versatility, making it one of the most popular languages for beginners and professionals and a strong ecosystem explain why it continues to grow in popularity for both learning and professional development.

2. What is an interpreter in Python?

>> An interpreter in Python is a software program that reads, processes, and executes Python code line by line, converting it from human-readable instructions into a form the computer can understand and run instantly.

3. What are pre-defined keywords in Python?

>> Pre-defined keywords in Python are special reserved words that form the foundation of the language’s syntax and have specific meanings, so they cannot be used as variable, function, or class names. 

4. Can keywords be used as variable names?

>> No, keywords cannot be used as variable names in Python. They are reserved words that have special meaning in the language's syntax, so attempting to use a keyword as a variable name results in a syntax error. 

5. What is mutability in Python?

>> Mutability in Python refers to whether an object’s value can be changed after it is created. If an object is mutable, its value or state can be directly modified without creating a new object. 

6. Why are lists mutable, but tuples are immutable?

>> Lists are mutable in Python because they are designed to be dynamic collections that can be modified after creation. This means you can add, remove, or change elements within a list without creating a new list. On the other hand, tuples are immutable, because once they are created, their contents cannot be changed. This immutability makes tuples suitable for fixed data collections that should remain constant throughout the program.

7. What is the difference between “==” and “is” operators in Python?

>> The difference between the == and is operators in Python lies in what they compare:
 (i) The “==” operator checks for equality of values. It evaluates whether the values of two objects are the same, regardless of whether they occupy the same memory location or not. 

(ii) The “is” operator checks for identity. It determines whether two variables point to the exact same object in memory

8. What are logical operators in Python?

>> Logical operators in Python are used to combine or modify conditional statements that evaluate to True or False. The three main logical operators are and, or, and not.
List of Logical operators:
 - and: Returns True if both operands are true.
 - or: Returns True if at least one operand is true.
 - not: Returns True if the operand is false; it negates the boolean value.

9. What is type casting in Python?

>> Type casting in Python is the process of converting a variable from one data type to another. It allows a programmer to change the type of data explicitly or implicitly to perform different operations or use the data in a different context.

10. What is the difference between implicit and explicit type casting?

>>The difference between implicit and explicit type casting in Python: 
 (i) Implicit Type Casting:
 - Also called automatic type conversion.
 - Performed automatically by Python without programmer intervention.
 - It ensures smooth operations by automatically converting types when needed.

 (ii) Explicit Type Casting:
 - Also called type casting or type conversion.
 - Performed manually by the programmer using built-in functions like int(), float(), str(), etc.
 - The programmer explicitly specifies the data type to convert to, providing full control over the conversion process.
 - Useful when strict control or clarity is needed in how data types are converted

11. What is the purpose of conditional statements in Python? 

>> The purpose of conditional statements in Python is to control the flow of program execution by making decisions based on whether specified conditions evaluate to True or False. They allow a program to execute certain blocks of code only when particular conditions are met, enabling dynamic behavior and decision-making.

12. How does the elif statement work?

>> The elif statement in Python stands for "else if" and is used to check multiple conditions sequentially after an initial if condition. It works by evaluating each condition in order, and as soon as one if or elif condition is true, its associated block of code runs, and the rest are skipped.

How elif works:
 - If the first if condition is False, Python checks the elif condition(s).
 - Only the block corresponding to the first True condition among if and elif statements executes.
 - If none of the if or elif conditions are true, the optional else block executes.
 - This allows for cleaner, readable code when testing multiple conditions without deeply nested if-else statements.

13. What is the difference between for and while loops?

>> (i) For Loop
 - Used to iterate over a sequence (like a list, tuple, string, or range) when the number of iterations is known or fixed in advance.
 - The loop variable takes each item from the sequence one by one automatically.
 - Ideal for iterating through elements of a collection or sequence.

  (ii) While Loop
 - Executes a block of code repeatedly as long as a given condition is True.
 - The number of iterations is not known in advance and depends on the condition being true.
 - Useful for situations requiring looping until a dynamic or complex condition changes.

14. Describe a scenario where a while loop is more suitable than a for loop?

>> A while loop is more suitable than a for loop when the number of iterations is unknown beforehand and the loop needs to continue executing until a specific condition changes dynamically. It is ideal when the termination condition depends on something that happens within the loop during execution rather than iterating over a fixed sequence.
